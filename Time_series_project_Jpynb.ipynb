```python
df1= df.loc[df['Category']=='Office Supplies']
df1.head()
```

```python
#print min and max order date
print(df1['Order Date'].min())
print(df1['Order Date'].max())
```

```python
#drop unwanted column
df1 = df1.drop(['OrderNumber',	'WarehouseCode',	'Ship Date',	'Category',	'Sales Channel',	'CurrencyCode',	'_SalesTeamID','_CustomerID',	'_StoreID',	'_ProductID','Quantity','Discount',	'Profit'], axis=1)
```

```python
df1= df1.groupby('Order Date')['Sales'].sum().reset_index()
df1.head()
```

```python
df1['Order Date'] = pd.to_datetime(df1['Order Date'], format='%d-%m-%Y')
df1= df1.set_index('Order Date')
df1.head()
```

```python
df1.info()
```

```python
df1.plot()
```

```python
#create new DataFrame
monthly_sales = pd.DataFrame()

monthly_sales['Sales'] = df1['Sales'].resample('MS').mean()
plt.plot(monthly_sales.index, monthly_sales.Sales, linewidth=3)
```

```python
import statsmodels.api as sm
```

```python
decomposition = sm.tsa.seasonal_decompose(monthly_sales['Sales'], model='additive')
fig = decomposition.plot()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
#rolling statistics
moving_avg = monthly_sales['Sales'].rolling(12).mean()
moving_std= monthly_sales['Sales'].rolling(12).std()
```

```python
#Plot rolling statistics:
plt.plot(monthly_sales['Sales'], color='blue',label='Original') # Plot the 'Sales' column
plt.plot(moving_avg, color='red', label='Rolling Mean')
plt.plot(moving_std, color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

```python
from statsmodels.tsa.stattools import adfuller
print ('Results of Dickey-Fuller Test:')
# Pass only the numerical 'Sales' column to adfuller
dftest = adfuller(monthly_sales['Sales'], autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])

for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print (dfoutput)
```

```python
#train test split
train = monthly_sales[:int(0.8*(len(monthly_sales)))]
test = monthly_sales[int(0.8*(len(monthly_sales))):]
```

```python
auto_arima(monthly_sales,test='adf',       # use adftest to find optimal 'd'
                       # maximum p and q
                                 # frequency of series
                       # let model determine 'd'
                      seasonal=True,   # No Seasonality


                      trace=True,
                      error_action='ignore',
                      suppress_warnings=True,
                      stepwise=True)
```

```python
#arima model
model=ARIMA(train, order=(2,1,2)).fit()
model.summary()
```

```python
pred= model.predict(start=len(train), end=(len(monthly_sales)-1),dynamic=True)
pred
```

```python
test
```

```python
train.plot(legend=True, label='Train', figsize=(10,6))

test.plot(legend=True, label= 'Test')

pred.plot(legend=True, label='ARIMA prediction')
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
mape=mean_absolute_error(test,pred)
mse=mean_squared_error(test,pred)
r2=r2_score(test,pred)
print('MAPE:',mape)
print('MSE:',mse)
print('R2:',r2)
```

```python
import statsmodels.api as sm
model1=sm.tsa.statespace.SARIMAX(train,order=(1, 1, 1),seasonal_order=(1,1,1,12))

results=model1.fit()
results.summary()
```

```python
pre=results.predict(start= len(train), end= (len(monthly_sales)-1),dynamic=True)
pre
```

```python
#plot train test pre data
train.plot(legend=True, label='Train', figsize=(10,6))

test.plot(legend=True, label= 'Test')

pre.plot(legend=True, label='SARIMA prediction')
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
mape=mean_absolute_error(test,pre)
mse=mean_squared_error(test,pre)
r2=r2_score(test,pre)
print('MAPE:',mape)
print('MSE:',mse)
print('R2:',r2)
```
